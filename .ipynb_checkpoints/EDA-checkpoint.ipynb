first step read the dataset and filter the rows depends on the visa class, then remove the extra columns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import math

pd.pandas.set_option('display.max_columns', None)

dataset = pd.read_excel("dataset_min.xlsx")

print("dataset shape is: "+str(dataset.shape))
print("remove rows not related to H-1B visa class and remove the extra columns manually")

not_H1B_visa = []
for i in range(len(dataset["VISA_CLASS"])):
    if dataset["VISA_CLASS"][i] != "H-1B":
        not_H1B_visa.append(i)

dataset = dataset.drop(not_H1B_visa)

#features_list = ['CASE_STATUS', 'CASE_SUBMITTED', 'JOB_TITLE', 'SOC_CODE', 'SOC_TITLE', 'FULL_TIME_POSITION', 'PERIOD_OF_EMPLOYMENT_START_DATE', 'PERIOD_OF_EMPLOYMENT_END_DATE', 'TOTAL_WORKER_POSITIONS', 'NEW_EMPLOYMENT', 'CONTINUED_EMPLOYMENT', 'CHANGE_PREVIOUS_EMPLOYMENT', 'NEW_CONCURRENT_EMPLOYMENT', 'CHANGE_EMPLOYER', 'AMENDED_PETITION', 'EMPLOYER_NAME', 'EMPLOYER_BUSINESS_DBA', 'EMPLOYER_ADDRESS1', 'EMPLOYER_CITY', 'EMPLOYER_STATE', 'EMPLOYER_POSTAL_CODE', 'EMPLOYER_COUNTRY', 'NAICS_CODE', 'AGENT_REPRESENTING_EMPLOYER', 'AGENT_ATTORNEY_CITY', 'AGENT_ATTORNEY_STATE', 'AGENT_ATTORNEY_POSTAL_CODE', 'AGENT_ATTORNEY_COUNTRY', 'STATE_OF_HIGHEST_COURT', 'WORKSITE_WORKERS_1', 'SECONDARY_ENTITY_1', 'SECONDARY_ENTITY_BUSINESS_NAME_1', 'WORKSITE_CITY_1', 'WORKSITE_COUNTY_1', 'WORKSITE_STATE_1', 'WORKSITE_POSTAL_CODE_1', 'WAGE_RATE_OF_PAY_FROM_1', 'WAGE_RATE_OF_PAY_TO_1', 'WAGE_UNIT_OF_PAY_1', 'PREVAILING_WAGE_1', 'PW_UNIT_OF_PAY_1', 'PW_WAGE_LEVEL_1', 'PW_OES_YEAR_1', 'PW_OTHER_SOURCE_1', 'PW_NON-OES_YEAR_1', 'WORKSITE_WORKERS_2', 'SECONDARY_ENTITY_2', 'SECONDARY_ENTITY_BUSINESS_NAME_2', 'WORKSITE_CITY_2', 'WORKSITE_COUNTY_2', 'WORKSITE_STATE_2', 'WORKSITE_POSTAL_CODE_2', 'WAGE_RATE_OF_PAY_FROM_2', 'WAGE_RATE_OF_PAY_TO_2', 'WAGE_UNIT_OF_PAY_2', 'PREVAILING_WAGE_2', 'PW_UNIT_OF_PAY_2', 'PW_WAGE_LEVEL_2', 'PW_OES_YEAR_2', 'PW_OTHER_SOURCE_2', 'PW_NON-OES_YEAR_2', 'PW_SURVEY_PUBLISHER_2', 'H-1B_DEPENDENT', 'WILLFUL_VIOLATOR', 'SUPPORT_H1B', 'STATUTORY_BASIS', 'PUBLIC_DISCLOSURE']
features_list = ['CASE_STATUS', 'EMPLOYER_NAME', 'SECONDARY_ENTITY_1', 'AGENT_REPRESENTING_EMPLOYER', 'JOB_TITLE', 'SOC_TITLE', 'SOC_CODE', 'NAICS_CODE', 'CONTINUED_EMPLOYMENT', 'CHANGE_PREVIOUS_EMPLOYMENT', 'NEW_CONCURRENT_EMPLOYMENT', 'CHANGE_EMPLOYER', 'AMENDED_PETITION', 'FULL_TIME_POSITION', 'H-1B_DEPENDENT', 'SUPPORT_H1B', 'WILLFUL_VIOLATOR', 'WAGE_RATE_OF_PAY_FROM_1', 'WAGE_UNIT_OF_PAY_1', 'TOTAL_WORKER_POSITIONS']
other_features_list = []

for c in dataset.columns:
    if c not in features_list:
        other_features_list.append(c)

dataset = dataset.drop(columns=other_features_list)
print("dataset shape is: "+str(dataset.shape))
```

take a look at empty cells, and determine the value should be fill

```python
row_num = len(dataset)
for d in dataset.columns:
    sum = dataset[d].isnull().sum()
    missing_rate = round(sum/row_num*100, 2)
    if sum > 0:
        print("{0:35}: {1:6}% missing, {2:6} unique values".format(str(d), str(missing_rate), str(dataset[d].unique())))
dataset.head()
```

we can assume that the missing values in those features are 'N'

```python
dataset_1 = dataset.copy()
dataset_1['AGENT_REPRESENTING_EMPLOYER'].fillna("N", inplace = True)
dataset_1['SECONDARY_ENTITY_1'].fillna("N", inplace = True)
dataset_1['H-1B_DEPENDENT'].fillna("N", inplace = True)
dataset_1['WILLFUL_VIOLATOR'].fillna("N", inplace = True)
dataset_1['SUPPORT_H1B'].fillna("N", inplace = True)

# make sure if there is any remaining empty values
row_num = len(dataset_1)
for d in dataset_1.columns:
    sum = dataset_1[d].isnull().sum()
    missing_rate = round(sum/row_num*100, 2)
    if sum > 0:
        print("{0:35}: {1:6}% missing, {2:6} unique values".format(str(d), str(missing_rate), str(dataset[d].unique())))

dataset_1.head()
```

now we decide which of the features are categorical or numerical
let us look at number of unique values for each feature

```python
for d in dataset_1.columns:
    unique_num = len(dataset_1[d].unique())
    print("{0:35}: {1:6}".format(str(d), str(unique_num)))

dataset_1.head()
```

we can see that only TOTAL_WORKER_POSITIONS and WAGE_RATE_OF_PAY_FROM_1 are numerical values

and the folloing features are boolean values:
FULL_TIME_POSITION, CONTINUED_EMPLOYMENT, CHANGE_PREVIOUS_EMPLOYMENT, NEW_CONCURRENT_EMPLOYMENT, CHANGE_EMPLOYER, AMENDED_PETITION, AGENT_REPRESENTING_EMPLOYER, SECONDARY_ENTITY_1, H-1B_DEPENDENT, WILLFUL_VIOLATOR, SUPPORT_H1B.

and the others are normal categorical values.

for the boolean features that have more than two unique values we have to consider 0 as 0 and any other value as 1, and this case presents in the following features:
CONTINUED_EMPLOYMENT, CHANGE_PREVIOUS_EMPLOYMENT, NEW_CONCURRENT_EMPLOYMENT, CHANGE_EMPLOYER, AMENDED_PETITION.

```python
dataset_2 = dataset_1.copy()

boolean_features_list = ['CONTINUED_EMPLOYMENT', 'CHANGE_PREVIOUS_EMPLOYMENT', 'NEW_CONCURRENT_EMPLOYMENT', 'CHANGE_EMPLOYER', 'AMENDED_PETITION']

for f in boolean_features_list:
    previous_num = len(dataset_2[f].unique())
    for key, value in dataset_2[f].iteritems(): 
        if value != 0:
            dataset_2.at[key,f] = 1
    print("number of unique values for {0:30}: was {1:3}, became {2:3}".format(f, previous_num, str(len(dataset_2[f].unique()))))
```

for the other boolean features we have values labeled as 'N' and 'Y', we have to convert them into 0 and 1

```python
dataset_3 = dataset_2.copy()

boolean_features_list = ['FULL_TIME_POSITION', 'AGENT_REPRESENTING_EMPLOYER', 'SECONDARY_ENTITY_1', 'H-1B_DEPENDENT', 'WILLFUL_VIOLATOR', 'SUPPORT_H1B']

for f in boolean_features_list:
    previous_num = len(dataset_3[f].unique())
    for key, value in dataset_3[f].iteritems(): 
        if value == 'N':
            dataset_3.at[key,f] = 0
        else:
            dataset_3.at[key,f] = 1
dataset_3.head()
```

now we need to make WAGE_RATE_OF_PAY_FROM_1 for all entries in the same scale, using WAGE_UNIT_OF_PAY_1 and FULL_TIME_POSITION we can determin the common rate as per hour

```python
dataset_4 = dataset_3.copy()

for key, value in dataset_4['WAGE_UNIT_OF_PAY_1'].iteritems():
    if value == 'Year':
        if dataset_4.at[key, 'FULL_TIME_POSITION']:
            dataset_4.at[key, 'WAGE_RATE_OF_PAY_FROM_1'] /= (40 * 52)
        else:
            dataset_4.at[key, 'WAGE_RATE_OF_PAY_FROM_1'] /= (25 * 52)
    elif value == 'Month':
        if dataset_4.at[key, 'FULL_TIME_POSITION']:
            dataset_4.at[key, 'WAGE_RATE_OF_PAY_FROM_1'] /= 176
        else:
            dataset_4.at[key, 'WAGE_RATE_OF_PAY_FROM_1'] /= 110
    elif value == 'Bi-Weekly':
        if dataset_4.at[key, 'FULL_TIME_POSITION']:
            dataset_4.at[key, 'WAGE_RATE_OF_PAY_FROM_1'] /= 40
        else:
            dataset_4.at[key, 'WAGE_RATE_OF_PAY_FROM_1'] /= 25
```
